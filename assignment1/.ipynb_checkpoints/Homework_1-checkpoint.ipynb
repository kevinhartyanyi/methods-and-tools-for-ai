**1. [1 point]** Write the `cylinder_volume` function that takes an array of shape Nx2, where the first column is the radius, the second is the length of the cylinder. The function calculates the volume of the cylinder.

```python
def cylinder_volume(a):    
    return np.pi * a[:, 1] * a[:, 0]**2 

assert np.abs(cylinder_volume (np.array([[1,1]])) - np.array(3.14159265))<1e-5
assert np.all(np.abs(cylinder_volume (np.array([[0,1], [2,2]])) - np.array([0, 25.132741]))<1e-5)
```

**2. [2 points]** Create the `moving_sum` function that calculates the moving sum on an array. Specifically, it takes two parameters: `a` and `w` where `w` is the window size. For a one dimensional array the return array must be:

`r[i] = a[i-w+1]+a[i-w+2]+...+a[i]`

Note the length of the output `r` is `w-1` elements shorter. If `a` has multiple dimensions, calculate the moving sum over the last axis.

```python
def moving_sum(a, w):
    ret = np.cumsum(a, axis=-1, dtype=float)
    ret[..., w:] = ret[..., w:] - ret[..., :-w]
    return ret[..., w - 1:]

assert np.all(moving_sum(np.array([1, 2, 3, 4, 5]), 2) == [3., 5., 7., 9.])
assert np.all(moving_sum(np.array([[1, 2, 3, 4, 5], [24, -5, 3, 4, 8]]), 3) == [[ 6.,  9., 12], [22.,  2., 15.]])
```

**3. [2 points]**  We have a video, recorded in 25 FPS. We also have a list of timestamps when a certain event occured. Your task is to return the index of frames, where at least 3 events happened. The input timestamps are in miliseconds.

```python
def at_least_3_events(timestamps):
    frames_per_second = 25
    one_frame_length_in_millisec = (1 / frames_per_second) * 1000
    mid = (timestamps / one_frame_length_in_millisec).astype(int)
    _, counts = np.unique(mid, return_counts=True)
    return np.where(counts >= 3)[0]
    

assert np.all(at_least_3_events(np.array([10, 20]))==[])
assert np.all(at_least_3_events(np.array([10, 20, 30, 40, 42, 81, 119, 90]))==[0, 2])
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('train.csv', sep=',')
df = df.sort_values(by=['x'])
```

```python
import matplotlib.pyplot as plt
plt.plot(df['x'], df['y']);
plt.title("Training Data");
```

```python
class LinearRegression:
    def __init__(self, lr=0.01, num_iter=10000, fit_intercept=True):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.train_theta = []
    
    def __add_intercept(self, X):
        # add a column of 1s
        intercept = np.ones((X.shape[0]))
        return np.stack((intercept, X), axis=1)
    
    def __loss(self, h, y):
        return np.mean(np.square(h - y))
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # weights initialization
        self.theta = np.zeros(X.shape[1])
        
        loss_vals = []
        t_theta = []
        for i in range(self.num_iter):
            h = np.dot(X, self.theta)
            gradient = 2*np.dot(X.T, h - y) / y.size
            self.theta -= self.lr * gradient
            
            t_theta.append(np.copy(self.theta))
            
            h = np.dot(X, self.theta)
            loss = self.__loss(h, y)
            loss_vals.append(loss)
        
        return loss_vals, t_theta
    
    def predict(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return np.dot(X, self.theta)
```

```python
def std_normalization(data):
    # params: data: ndarray(n_samples, n_features)
    data = np.asarray(data, dtype=np.float64)
    eps = 0.0000001
    x_mean = np.mean(data, axis=0)
    x_std = np.std(data, axis=0)
    x_norm = (data-x_mean) / (x_std + eps)
    
    return x_norm, x_mean, x_std
  
def std_normalize_with_given_meanstd(data, x_mean, x_std):
    eps = 0.0000001
    x_norm = (data-x_mean) / (x_std + eps)
    return x_norm
```

```python
X_train = df['x'].values
y_train = df['y'].values
```

```python
X_train_n, x_mean, x_std = std_normalization(X_train)
```

**4. [2 points]** First fit a simple linear regression on the data, using bias parameters as well. The target variable is in the `y` column. What is the loss on the dataset? Plot the results!

```python
trained_models = []

model = LinearRegression(lr=0.001, num_iter=10000)

loss, t_theta = model.fit(X_train_n, y_train)
trained_models.append(model)

plt.plot(loss)
plt.title("Training Loss")
plt.show();

preds_train = model.predict(X_train_n)
print("Training mean absolute error: " + str(np.abs(preds_train - y_train).mean()))
```

```python
def draw_comparison_plot(predicted_values, name="", gt=df['y'], x_cord=X_train_n):
    plt.plot(x_cord, gt, label="true");
    plt.plot(x_cord, predicted_values, label="predicted");
    plt.title(f"Comparison {name}");
    plt.legend(loc="lower right");
    plt.show()
```

**5. [4 points]** Now implement polynomial features in the linear regeression model. Try polynomials up until of degree 4. What is their loss? Which one looks the best?

```python
class PolynomialRegression:
    def __init__(self, lr=0.01, num_iter=10000, degree=1, fit_intercept=True):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.train_theta = []
        self.degree = degree
    
    def __add_intercept(self, X):
        # add a column of 1s
        intercept = np.ones((X.shape[0]))
        return np.stack((intercept, X), axis=1)
    
    def __loss(self, h, y):
        return np.mean(np.square(h - y))
    
    def __convert_X(self, X):
        return np.concatenate([X] + [X[:,1:]**i for i in range(2, self.degree+1)], axis=1)
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        # weights initialization        
        
        ####
        self.theta = np.zeros((X.shape[1] + (self.degree - 1)))
        X = self.__convert_X(X)
        ####
        
        loss_vals = []
        t_theta = []
        for i in range(self.num_iter):
            h = np.dot(X, self.theta)
            gradient = 2*np.dot(X.T, h - y) / y.size
            self.theta -= self.lr * gradient
            
            t_theta.append(np.copy(self.theta))
            
            h = np.dot(X, self.theta)
            loss = self.__loss(h, y)
            loss_vals.append(loss)
        
        return loss_vals, t_theta
    
    def predict(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        X = self.__convert_X(X)
    
        return np.dot(X, self.theta)
```

```python
for degree in range(2,5):
    model_poly = PolynomialRegression(lr=0.001, num_iter=10000, degree=degree)
    loss, t_theta = model_poly.fit(X_train_n, y_train)
    
    trained_models.append(model_poly)

    plt.plot(loss)
    plt.title(f"Training Loss with degree: {degree}")
    plt.show();

    preds_train_poly = model_poly.predict(X_train_n)
    print(f"Training mean absolute error with degree: {degree}: " + str(np.abs(preds_train_poly - y_train).mean()))
    draw_comparison_plot(preds_train_poly, f"with degree: {degree}")
```
